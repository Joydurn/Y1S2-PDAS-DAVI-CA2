# Importing Libraries and Data

```python
#filter to make our output cleaner
import plotly
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) 
#main dependencies
import numpy as np
import pandas as pd
#makes working environment cleaner
pd.reset_option('all')
pd.options.mode.chained_assignment = None 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path="data/og/"
concession2009DF=pd.read_excel(data_path+"2009_Concession.xlsx")
busData2009DF=pd.read_csv(data_path+"2009_BUS.csv")
mrtData2009DF=pd.read_csv(data_path+"2009_MRT.csv")

concessionDF=pd.read_csv(data_path+"concession_card_price.csv")
feederDF=pd.read_csv(data_path+"feeder_fare.csv")
mrtDF=pd.read_csv(data_path+"mrt_fare.csv")

ridershipOldDF=pd.read_csv(data_path+"ridership1995-2016.csv")
ridershipNewDF=pd.read_excel(data_path+"ridership2015-2020.xlsx")
```

```python
#Taking a look at our datasets before cleaning
display(concession2009DF)
display(busData2009DF)
display(mrtData2009DF)

display(concessionDF)
display(feederDF)
display(mrtDF)

display(ridershipOldDF)
display(ridershipNewDF)
```

# Dataframe Cleaning and Transforming

## Concession Cards

```python
#concession cleaning
#2020 data
#remove other cards other than polytechnic students because only comparing for SP
concessionDF1=concessionDF.loc[concessionDF['cardholders'].isin(['Polytechnic Student'])] #filter to just get polytechnic student prices
concessionDF1.rename(columns={'cardholders':'cardholder'},inplace=True, errors='raise') 

#2009 Data
concession2009DF1=concession2009DF.loc[concession2009DF['cardholder'].isin(['Polytechnic'])]
concession2009DF1.loc[0, 'cardholder'] = 'Polytechnic Student'

concessionDF1['year'] = 2020
concession2009DF1['year'] = 2009

#merging 2009 and 2020 data
concessDF=pd.concat([concession2009DF1,concessionDF1])
concessDF.reset_index(drop=True,inplace=True)
concessDF.rename(columns={'bus_price':'bus card','train_price':'train card','hybrid_price':'bus and train card'},inplace=True, errors='raise') 
concessDF.drop(['cardholder'], axis = 1, inplace = True) 
concessDF1=pd.melt(concessDF,id_vars=['year'],value_vars=['bus card','train card','bus and train card'],
                   value_name='price', var_name='type')
concessDF1['year'] = concessDF1['year'].astype(str)
concessDF1['price']=pd.to_numeric(concessDF1['price'], errors='coerce')
print("****************************************BEFORE CLEANING:***********************************************")
display(concessionDF)
display(concession2009DF)
print("****************************************AFTER CLEANING:***********************************************")
display(concessDF1)

#no null values
```

## Bus Fares

```python
feederDF1=feederDF.transpose() #transpose
#use first row as header 
new_header = feederDF1.iloc[0] #grab the first row for the header
feederDF1 = feederDF1[1:] #take the data less the header row
feederDF1.columns = new_header #set the header row as the feederDF1 header
feederDF1.reset_index(level=0, inplace=True)
feederDF1 = feederDF1.rename_axis(None, axis=1)
feederDF1.rename(columns={'index':'fare_type','Fare Per Ride (cent)':'fare'},inplace=True, errors='raise') 
feederDF2=feederDF1.loc[feederDF1['fare_type'].isin(['adult_card_fare_per_ride','student_card_fare_per_ride'])]
feederDF2.reset_index(drop=True, inplace=True)
feederDF2.at[0,'fare_type']='Adult 2020'
feederDF2.at[1,'fare_type']='Student 2020'
feederDF2['fare'] = pd.to_numeric(feederDF2['fare'], errors='coerce')
```

### Merging 2009 and 2020 Data

```python
#merging with 2009 data
busData2009DF1=busData2009DF
busData2009DF1.at[0,'fare type']='Adult 2009'
busData2009DF1.at[1,'fare type']='Student 2009'
busData2009DF1.rename(columns={'fare type':'fare_type'},inplace=True, errors='raise') 
#convert to cents
busData2009DF1['fare'] = pd.to_numeric(busData2009DF1['fare'], errors='raise')
busData2009DF1['fare']=busData2009DF1['fare'].apply(lambda x: x*100)
#concatenate dataframes
busDF=pd.concat([busData2009DF1,feederDF2])
busDF.reset_index(drop=True, inplace=True) #reset index
#split fare type column into Type and Year
busDF[['Type','Year']] = busDF.fare_type.str.split(" ", expand=True)
busDF.drop(['fare_type'], axis = 1, inplace = True) 
busDF = busDF.reindex(columns = ['Year','Type','fare'])
busDF['Year'] = pd.to_numeric(busDF['Year'], errors='coerce')
busDF['fare'] = pd.to_numeric(busDF['fare'], errors='coerce')
print("****************************************BEFORE CLEANING:***********************************************")
display(feederDF)
display(busData2009DF)
print("****************************************AFTER CLEANING:***********************************************")
display(busDF)
g=sns.histplot(x='fare' ,data=busDF)
plt.savefig('graphs/12.pdf') #export HD pdf
plt.show()
```

## MRT Fares

```python
mrtDF1=mrtDF.loc[mrtDF['fare_type'].isin(['Student card fare','Adult card fare'])]
mrtDF1.reset_index(drop=True, inplace=True)
```

```python
#make column names match each other
#add column that labels student as student DF
mrtDF1=mrtDF.loc[mrtDF['fare_type'].isin(['Student card fare','Adult card fare'])]
mrtDF1.reset_index(drop=True, inplace=True)
#handling null values
print("Rows with null values:")
display(mrtDF1[mrtDF1.isna().any(axis=1)])
#use forward filling because the fare values surrounding the null values are the same
mrtDF1=mrtDF1.fillna(method='ffill')
#renaming values
mrtDF1['fare_type'] = mrtDF1['fare_type'].str.replace('Adult card fare','Adult')
mrtDF1['fare_type'] = mrtDF1['fare_type'].str.replace('Student card fare','Student')
mrtDF1['applicable_time'] = mrtDF1['applicable_time'].str.replace('All other timings','After 7.45am')
mrtDF1['applicable_time'] = mrtDF1['applicable_time'].str.replace('All timings','After 7.45am')

#process distance column to be numbers
mrtDF1['distance'] = mrtDF1['distance'].str.replace('Up to 3.2 km','3.2')
mrtDF1= mrtDF1[mrtDF1.distance != 'Over 40.2 km'] #remove redundant rows
for index,row in mrtDF1.iterrows():
    if row["distance"]!='3.2':
      string=row["distance"].split(' - ')
      distance=string[1][:-3]
      mrtDF1["distance"][index]=distance
mrtDF1.rename(columns={'distance':'distance(km)','fare_per_ride':'Estimated Monthly Cost($)'},inplace=True, errors='raise') #monthly cost calculated later
#insert year column
mrtDF1.insert(0, 'year', np.full(152, 2020))
#convert distance to float type
mrtDF1['distance(km)'] = pd.to_numeric(mrtDF1['distance(km)'], errors='coerce')
mrtDF1.reset_index(drop=True,inplace=True)

#Fixing an issue with original dataset
print("Dataset has a mistake/problem. These rows below are duplicate entries with different costs. \n However, I figured out that the duplicate entries with lower costs are the ones that actually detail the fares for adult fares before 7.45am. Let's fix this by replacing it with 'Before 7.45am'. ")
mrtDF1.head(38)['applicable_time']=mrtDF1.head(38)['applicable_time'].str.replace('After 7.45am','Before 7.45am')
```

```python
#processing 2009 data
#process distance column to be numbers
mrtData2009DF['distance'] = mrtData2009DF['distance'].str.replace('Up to 3.2 km','3.2')
mrtData2009DF1= mrtData2009DF[mrtData2009DF.distance != 'Over 40.2 km'] #remove redundant rows
for index,row in mrtData2009DF1.iterrows():
    if row["distance"]!='3.2':
      string=row["distance"].split(' - ')
      distance=string[1][:-3]
      mrtData2009DF1["distance"][index]=distance
#change column name
mrtData2009DF1.rename(columns={'distance':'distance(km)'},inplace=True, errors='raise')
#convert string column to float
mrtData2009DF1['adult fare'] = pd.to_numeric(mrtData2009DF1['adult fare'], errors='coerce')
mrtData2009DF1['student fare'] = pd.to_numeric(mrtData2009DF1['student fare'], errors='coerce')
mrtData2009DF1['distance(km)'] = pd.to_numeric(mrtData2009DF1['distance(km)'], errors='coerce')
#due to the nature of this dataset, rows past the 21.2km mark for the distance is null, most likely due to the fact that the MRT lines were much shorter in 2009 compared to today. It would not be fair to compare the fares when these lines are not built yet, so we will eliminate these and only compare fares for shorter distances.
mrtData2009DF1=mrtData2009DF1.dropna()
#adjust for inflation so fare comparison is fair
#inflation rate from 2009 to 2021 (https://www.creatifwerks.com/singapore-inflation-calculator/)
#S$ 400.00 back in 2009 is worth Purchasing Power Value of S$ 476.54 in 2021
inflateRate=1.19135
mrtData2009DF1['adult fare']=mrtData2009DF1['adult fare'].apply(lambda x: round(x*100*inflateRate))
mrtData2009DF1['student fare']=mrtData2009DF1['student fare'].apply(lambda x: round(x*100*inflateRate))
#mrtData2009DF1.round({'adult fare': 1, 'student fare': 1})
#handling null values

#melting into fare type
mrtData2009DF1=pd.melt(mrtData2009DF1, id_vars=['distance(km)'], value_vars=['adult fare', 'student fare'], var_name='fare_type', value_name='Estimated Monthly Cost($)')
mrtData2009DF1 = mrtData2009DF1[['fare_type','distance(km)', 'Estimated Monthly Cost($)']] #monthly cost is calculated later
#insert year column
mrtData2009DF1.insert(0, 'year', np.full(36, 2009))
#renaming values
mrtData2009DF1['fare_type'] = mrtData2009DF1['fare_type'].str.replace('adult fare','Adult')
mrtData2009DF1['fare_type'] = mrtData2009DF1['fare_type'].str.replace('student fare','Student')

#applicable_time column is absent in 2009 due to fare not accomodating this at the time. i will insert the column so that it can match the other dataset when merging
mrtData2009DF1=pd.concat([mrtData2009DF1,mrtData2009DF1])
array1=np.full(36,'Before 7.45am')
array2=np.full(36,'After 7.45am')
timeArray=np.concatenate([array1,array2])
mrtData2009DF1.insert(0, 'applicable_time', timeArray)
```

### Merging 2009 and 2020 Data

```python
#merging
mrtDF2=pd.concat([mrtData2009DF1,mrtDF1])
mrtDF2.reset_index(drop=True, inplace=True)
#pivoting and melting
#mrtDF2=pd.pivot_table(mrtDF2,index=['applicable_time','distance(km)'],columns=['fare_type','year'],values=['Estimated Monthly Cost($)'])
print("Rows with null values after merging")
display(mrtDF2[mrtDF2.isna().any(axis=1)])

#Creating 'number of weekdays' column
newDF = pd.DataFrame(np.repeat(mrtDF2.values, 5, axis=0)) #repeat every row 5 times
newDF.columns = mrtDF2.columns
display(newDF)
#Make new column
daysArray=np.tile([1,2,3,4,5],(224,1)).reshape(1120)
newDF.insert(4, 'Weekdays Travelled', daysArray) 
mrtDF3=newDF
#these null values are caused by 2009 data not having certain data points due to different fare structure at the time. We will keep in mind these discrepancies
```

```python
#estimating totals for amounts used by travellers in a month
def estimateCost(type,year,fare,weekdays):
    busFareSeries=(busDF.query(f'Year=={year}')['fare'])
    if busFareSeries.count()>0:
        if type=='Student':
            busFare=busFareSeries.iloc[1]
        else:
            busFare=busFareSeries.iloc[0]
    else:
        busFare=0
    if year==2020:
        if fare>busFare:
            fare=fare
        else:
            fare=busFare
    elif year==2009:
        fare=fare+busFare
    dailyFare=fare*2
    weeklyFare=dailyFare*weekdays
    weeks=4
    monthlyFare=weeks*weeklyFare/100
    return monthlyFare

for index, row in mrtDF3.iterrows():
    currentFare=row['Estimated Monthly Cost($)']
    newFare=estimateCost(row['fare_type'],row['year'],row['Estimated Monthly Cost($)'],row['Weekdays Travelled'])
    mrtDF3.at[index, 'Estimated Monthly Cost($)'] = round(newFare,2)

mrtDF3['year'] = mrtDF3['year'].astype(str)
#merge time and card type together
mrtDF3["type"] = mrtDF3["fare_type"] + ' ' +mrtDF3["applicable_time"]
mrtDF3.drop(['applicable_time','fare_type'], axis = 1, inplace = True) 
#reorder the columns
mrtDF3 = mrtDF3.reindex(columns = ['year','type','distance(km)','Weekdays Travelled','Estimated Monthly Cost($)'])
```

```python
print("****************************************BEFORE CLEANING:***********************************************")
display(mrtDF)
display(mrtData2009DF)
print("****************************************AFTER CLEANING:***********************************************")
display(mrtDF3)
```
